##Formatting Function

```python
#%%
def format_output(text, max_line_length=80):
    lines = []
    current_line = ""

    for word in text.split():
        if len(current_line) + len(word) + 1 <= max_line_length:
            current_line += word + " "
        else:
            lines.append(current_line.strip())
            current_line = word + " "

    if current_line:
        lines.append(current_line.strip())

    return '\n'.join(lines)
```
